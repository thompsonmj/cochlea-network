```python
def load_data(filepath,sheet):
    #df = pd.read_excel(open(filepath,'rb'),sheet_name=sheet)
    df = pd.read_excel(filepath,sheet_name=sheet)
    
    return df

df = load_data('published-data/development-of-the-mouse-and-human-cochlea-at-single-cell-resolution/media-2.xlsx',
               'epi_data_matrix')

# Rename the first column to something useful
df.rename(columns={'Unnamed: 0':'Gene'},inplace=True)

# Display the first 10 rows
df.head(10) # or tail(10) will show the bottom 3   
```

```python
# Isolate groups making up the cochlea floor
# Use list comprehension to join suffix to groups
suffix = '_Avg_diff'
all_groups = ['10.1.1','10.1.2','10.1.3','10.2.0','10.2.1','10.3.0','10.3.1',\
             '10.3.2','10.4','10.5','22.0','22.1','22.2.0','22.2.1','22.3',\
             '24.0','24.1']
early_groups = ['24.0','24.1','10.0','10.1.0','10.1.1','10.4','10.2.1','10.3.0']
cycling_groups = ['10.0','24.0','24.1']

all_diffs = [group + suffix for group in all_groups]
early_diffs = [group + suffix for group in early_groups]
cycling_diffs = [group + suffix for group in cycling_groups]

# Rename the first column to something useful
df.rename(columns={'Unnamed: 0':'Gene'},inplace=True)
# From Figure 3, select the clusters in the floor corresponding to earliest timepoints

# Get just the columns we're interested (average differential expression)
cols = all_diffs
cols.insert(0,'Gene')
df_alldiffs = df[cols]

cols = early_diffs
cols.insert(0,'Gene')
df_earlydiffs = df[cols]

cols = cycling_diffs
cols.insert(0,'Gene')
df_cyclingdiffs = df[cols]
```
